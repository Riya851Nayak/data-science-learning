```python
import pandas as pd

data = {
    "Name": ["Riya", "Aman", "Neha", "Rohit", "Pooja"],
    "Age": [20, 21, 19, 22, 20],
    "Gender": ["Female", "Male", "Female", "Male", "Female"],
    "Marks": [85, 78, 92, 88, 90]
}

df=pd.DataFrame(data)
print(df)
```

Histogram

```python
plt.hist(df["Marks"])
plt.xlabel("Marks")
plt.ylabel("Count")
plt.title("Histogram of Marks")
plt.show()
```

Bar chart


```python
df.groupby("Gender")["Marks"].mean().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average Marks")
plt.title("Average Marks in Gender")
plt.show()
```

Line Chart

```python
plt.plot(df["Marks"])
plt.xlabel("Index")
plt.ylabel("Marks")
plt.title("Marks in trend")
plt.show()
```
