```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month" : ["Jan","Feb","Mar","Apr","May"],
    "Sales" : [2000,2500,1800,3000,2800]
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["Month"],df["Sales"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly sales trend")
plt.show()
```

```python
plt.bar(df["Month"],df["Sales"])
plt.title("Monthly sales trend")
plt.show()
```
